# Compute performance metrics for the given Y and Y_score without sklearn

```python
import numpy as np
import pandas as pd

#reading csv file and appending it to pandas data frame
data=pd.read_csv('5_a.csv')

# function to predict y_pred value on the basis of threshold
def y_predict(row,i):
    if row['proba']>i:
        return 1
    else:
        return 0
# adding one new column (Y-pred) in dataframe by calling for threshold 0.5
data['Y_Pred'] = data.apply (lambda row: y_predict(row,0.5), axis=1)

#TN,FP,FN,TP
seriesObj = data.apply(lambda x: True if ((x['y']==1) and (x['Y_Pred']==1)) else False, axis=1)
seriesObj_1 = data.apply(lambda x: True if ((x['y']==0) and (x['Y_Pred']==1)) else False, axis=1)
seriesObj_2 = data.apply(lambda x: True if ((x['y']==1) and (x['Y_Pred']==0)) else False, axis=1)
seriesObj_3 = data.apply(lambda x: True if ((x['y']==0) and (x['Y_Pred']==0)) else False, axis=1)

TP = len(seriesObj[seriesObj == True].index)
FP = len(seriesObj_1[seriesObj_1 == True].index)
FN = len(seriesObj_2[seriesObj_2 == True].index)
TN = len(seriesObj_3[seriesObj_3 == True].index)

# calculating confusion matrix using TN,FP,FN,TP values

Confusion_matrix=[[TP,FP],[FN,TN]]
print("Confusion_matrix for the given threshold 0.5 is " +str(Confusion_matrix))

# calculating F1_score using precision and recall

Precision=(TP/(TP+FP))
Recall=(TP/(TP+FN))
F1_score=((2*Precision*Recall)/(Precision+Recall))
print("F1_score for the given threshold 0.5 is " +str(F1_score))

#calculating accuracy using TN,FP,FN,TP values

Accuracy=((TP+TN)/(TP+TN+FN+FP))
print("Accuracy for the given threshold 0.5 is " +str(Accuracy))

#computing Auc scores

#creating thresholds between 0.01 to 0.99
threshold_list=(np.linspace(0.01,0.99,99))

#creating two empty array for TPR and FPR
tpr_array=[]
fpr_array=[]

#creating empty list for accuracy for each threshold list
Accuracy_lst=[]

cols=['probability', 'Accuracy']
#f1 = pd.DataFrame({"a":[],"b":[]})

data=data.drop(['Y_Pred'],axis=1)
data= data.sort_values(['proba'])


for i in threshold_list:
    
    # adding one new column (Y-pred) in dataframe by calling for threshold 0.5
    
    data['Y_Pred'] = data.apply (lambda row: y_predict(row,i), axis=1)
    
    #TN,FP,FN,TP for each threshold value
    seriesObj = data.apply(lambda x: True if ((x['y']==1) and (x['Y_Pred']==1)) else False, axis=1)
    seriesObj_1 = data.apply(lambda x: True if ((x['y']==0) and (x['Y_Pred']==1)) else False, axis=1)
    seriesObj_2 = data.apply(lambda x: True if ((x['y']==1) and (x['Y_Pred']==0)) else False, axis=1)
    seriesObj_3 = data.apply(lambda x: True if ((x['y']==0) and (x['Y_Pred']==0)) else False, axis=1)

    TP = len(seriesObj[seriesObj == True].index)
    FP = len(seriesObj_1[seriesObj_1 == True].index)
    FN = len(seriesObj_2[seriesObj_2 == True].index)
    TN = len(seriesObj_3[seriesObj_3 == True].index)
    
    
    #calculating tpr and fpr value and appending it to respective array
    tpr_array.append(TP/(TP+FN))
    fpr_array.append(FP/(FP+TN))
    
    #calculating accuracy for each threshold and append it to respective array
    Accuracy = ((TP+TN)/(TP+TN+FP+FN))
    
    Accuracy_lst.append([i,Accuracy])
    data=data.drop(['Y_Pred'],axis=1)

tpr_array.sort()
fpr_array.sort()

# calculating Auc value using np.trapz
Auc=np.trapz(tpr_array, fpr_array)
print("Auc scores is " +str(Auc))

Accuracy_df1 = pd.DataFrame(Accuracy_lst, columns=cols)

# other than these two you should not import any other packages
```
